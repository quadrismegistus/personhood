#### Does it get harder to be inhuman for an abstract noun?

```python
from lit import tools,os
def find_word(word):
    for dx in tools.readgen(HLI_FN):
        if word==dx['word']:
            yield dx
```

```python
find_word('joy').next()
```

```python
import pandas as pd
def find_word_db(word):
    return pd.DataFrame(list(find_word(word)))
```

```python

HLI_DIR_CROSSTABS = 'data_booknlp/data_by_quarter_century__crosstabs__rel_lemma'
def get_crosstab(period='1900-1924'):
    files = [fn for fn in os.listdir(HLI_DIR_CROSSTABS) if period in fn and fn.endswith('.txt')]
    if not files: return
    ifnfn=os.path.join(HLI_DIR_CROSSTABS,files[0])
    df=pd.read_csv(ifnfn,sep='\t',encoding='utf-8',quoting=3,error_bad_lines=False)
    return df.set_index('word')
    
```

```python
df_C18=get_crosstab('1700-1724')
df_C19=get_crosstab('1800-1824')
df_C20=get_crosstab('1900-1924')
```
